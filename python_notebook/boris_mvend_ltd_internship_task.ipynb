# Filling missing values

```python
# Fill in missing values
# For cat cols and date cols fill in with mode and for num cols fill in with mean
for col in trans_his_data.columns:
  if col in trans_date_cols+trans_cat_cols:
    trans_his_data[col] = trans_his_data[col].fillna(trans_his_data[col].mode()[0])
  elif col in trans_num_cols:
    trans_his_data[col] = trans_his_data[col].fillna(trans_his_data[col].mean())

# Confirm that there aren't any missing values
trans_his_data[trans_his_data.columns].isna().sum()
```

```python
# Fill in missing values
# For cat cols and date cols fill in with mode and for num cols fill in with mean

for col in sub_orders_data.columns:
  if col in sub_date_cols+sub_cat_cols:
    sub_orders_data[col] = sub_orders_data[col].fillna(sub_orders_data[col].mode()[0])
  elif col in sub_num_cols:
    sub_orders_data[col] = sub_orders_data[col].fillna(sub_orders_data[col].mean())

# Confirm that there aren't any missing values
sub_orders_data[sub_orders_data.columns].isna().sum()
```

# Data summary and merge to one dataframe

```python
# grouping by transaction_id
sub_orders_data_summary = []
for index, group in sub_orders_data.groupby("transaction_id"):
  record = [group['transaction_id'].mode()[0], group['transaction_date'].mode()[0], group['transaction_type'].mode()[0],  
            group['account_number'].mode()[0],   group['provider_id'].mode()[0], group['product_id'].sum(), group['duration'].sum(),
            group['product_price'].sum(),	 group['provider_price'].sum()]
  sub_orders_data_summary.append(record)

cols = ['transaction_id', 'transaction_date', 'transaction_type',
       'account_number', 'provider_id', 'product_id', 'duration',
       'product_price', 'provider_price']
sub_orders_data_summary = pd.DataFrame(data=sub_orders_data_summary, columns=cols)
```

# Teller Prediction (Classification Problem) Split Data Set and Build Model

```python
print("Are there missing values in the all_data? : ", all_data.isnull().any().any())
print((all_data.isnull().sum() / all_data.shape[0]  * 100))
```

```python
# Date extract
for col in trans_date_cols:
  for date_feature in ['year', 'month', 'day', 'quarter']:
    all_data[col+date_feature] = getattr(all_data[col].dt, date_feature)
  all_data.drop(columns=col,axis=1, inplace=True)
  
```

```python
# map features to numbers
from sklearn import preprocessing
cat_features.remove('teller_id')
label_encoder = preprocessing.LabelEncoder()
all_label_encoder_features = cat_features
for feature in all_label_encoder_features:
  label_encoder.fit(all_data[feature])
  all_data[feature] = label_encoder.transform(all_data[feature])

all_data.head()
```

```python
all_data.info()
```

```python
# Select main columns to be used in training
main_cols = all_data.columns.difference(trans_date_cols+['teller_id'])
X = all_data[main_cols]
y = all_data.teller_id

# Split data into train and test 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score


#Create array of models to evaluate
models = []

models.append(('Ridge', RidgeClassifier()))
models.append(('Bag_Re', BaggingClassifier()))
models.append(('RandomForest', RandomForestClassifier(n_estimators=10)))
models.append(('ExtraTreesRegressor', ExtraTreesClassifier(n_estimators=10)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('CBR', CatBoostClassifier(logging_level='Silent', cat_features=cat_features)))
models.append(('LB', LGBMClassifier(cat_features=cat_features)))
models.append(('XG', XGBClassifier(cat_features=cat_features)))



# Evaluations
results = []
names = []
scoring = []

from sklearn.metrics import classification_report
target_names=['class 0', 'class 1']

for name, model in models:
    # Fit the model
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    score = accuracy_score(y_test.astype(int), y_pred.astype(int))
    results.append(score)
    names.append(name)
    
    msg = "%s: %f " % (name, score)
    print(msg)

    
```

# Hyper-parameter Tuning and Building Final Model

```python
from sklearn.model_selection import GridSearchCV
clf = CatBoostClassifier(cat_features=cat_features)
params = {'learning_rate':[0.01, 0.015, 0.02, 0.025, 0.03],
          'iterations': [500, 1000, 1500],
          'depth': [4, 6, 8],
          'loss_function': ['Logloss'],
          'l2_leaf_reg': np.logspace(-20, -19, 3),
          'leaf_estimation_iterations': [8, 10],
          'logging_level':['Silent'],
          'random_seed': [42]
         }
clf_grid = GridSearchCV(estimator=clf, param_grid=params, scoring='accuracy', cv=5, verbose=3)
clf_grid.fit(X_train, y_train)
best_param = clf_grid.best_params_
best_param
```
